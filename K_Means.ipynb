```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
dataset = pd.read_csv(r'E:\Udemy\Machine Learning A-Z\Part 4 - Clustering\Section 24 - K-Means Clustering\Mall_Customers.csv')

X = dataset.iloc[:, [3,4]].values
X
```

```python
# Fitting dataset and Finding optimal K Clusters using Elbow method graph

from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11),wcss)
plt.title('Elbow method to find Optimal no. of Cluster')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Applying K-Means to dataset

kmean = KMeans(n_clusters = 5)
y_kmean = kmean.fit_predict(X)
y_kmean
```

```python
# Plotting graph

X = np.array(X)
plt.scatter(X[y_kmean == 0, 0], X[y_kmean == 0, 1], c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmean == 1, 0], X[y_kmean == 1, 1], c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmean == 2, 0], X[y_kmean == 2, 1], c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmean == 3, 0], X[y_kmean == 3, 1], c = 'purple', label = 'Cluster 4')
plt.scatter(X[y_kmean == 4, 0], X[y_kmean == 4, 1], c = 'pink', label = 'Cluster 5')
plt.scatter(kmean.cluster_centers_[:, 0], kmean.cluster_centers_[:, 1], c = 'yellow',s =300, label = 'Centroid', alpha =0.75)
plt.title('K-Means Clustering')
plt.xlabel('Age')
plt.ylabel('Salary in k$')
plt.legend()
plt.show()
```
